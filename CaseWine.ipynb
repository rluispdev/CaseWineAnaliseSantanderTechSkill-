3 - Utilizando a base de vinhos, responda:  exite alguma coluna com outliers? Indique qual mÃ©todo de detecÃ§Ã£o vocÃª utilizou, justificando seu caso.

```python

for col in df.columns:
    m,s = df[col].mean(), df[col].std()
    obs_eh_outlier = df[col].apply(lambda x: (x-m)/s).apply(lambda x: abs(x)> 3)
    obs_eh_outlier.value_counts()
    outliers = np.where(obs_eh_outlier)
    
    print(f"Os outliers sÃ£o: \n\n{outliers}")
    print(f"\n{'='*80}\n")
```
